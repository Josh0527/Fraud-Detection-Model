```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import requests
```

# Data Collection

```python
import kagglehub
import os

path = kagglehub.dataset_download("rohanrao/formula-1-world-championship-1950-2020")

print("Path to dataset files:", path)

#print file name

for filename in os.listdir(path):
    print(filename)
```

```python
#import files

df1 = pd.read_csv(path + "/circuits.csv", )
df2 = pd.read_csv(path + "/constructor_results.csv")
df3 = pd.read_csv(path + "/constructor_standings.csv")
df4 = pd.read_csv(path + "/constructors.csv")
df5 = pd.read_csv(path + "/driver_standings.csv")
df6 = pd.read_csv(path + "/drivers.csv")
df7 = pd.read_csv(path + "/lap_times.csv")
df8 = pd.read_csv(path + "/pit_stops.csv")
df9 = pd.read_csv(path + "/qualifying.csv")
df10 = pd.read_csv(path + "/races.csv")
df11 = pd.read_csv(path + "/results.csv")
df12 = pd.read_csv(path + "/seasons.csv")
df13 = pd.read_csv(path + "/status.csv")
df14 = pd.read_csv(path + "/sprint_results.csv")
```

```python
#create join in batches

batch_1 = pd.merge(df2, df3, on='raceId')
batch_1.head(3)
```

```python
#batch 2

batch_2 = pd.merge(df7, df8, on='raceId')
```

```python
#batch_3

batch_3 = pd.merge(df9, df10, on='raceId')
```

```python
#batch_4

batch_4 = pd.merge(df11, df14, on='raceId')
```

```python
#batch_5

batch_5 = pd.merge(df5, df6, on='driverId')
```

```python
#batch_3_new

batch_3_new = pd.merge(batch_3, df1, on='circuitId')
```

```python
batch_3_new.head(2)
```

```python
#batch_1_new

batch_1_new = pd.merge(batch_1, df4, left_on='constructorId_x', right_on='constructorId')
```

```python
batch_1_new.head(2)
```

# Power joins

```python
#covert each batch to csv

batch_1_new.to_csv("batch_1_new.csv", index=False)
batch_2.to_csv("batch_2.csv", index=False)
batch_3_new.to_csv("batch_3_new.csv", index=False)
batch_4.to_csv("batch_4.csv", index=False)
batch_5.to_csv("batch_5.csv", index=False)
```
